# scraping website

```python
url="https://www.flipkart.com/search?q=watch+for+women&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_5_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_3_5_na_na_na&as-pos=3&as-type=RECENT&suggestionId=watch+for+women&requestId=ac1ad346-848f-4631-b4c9-7649eed09993&as-searchtext=watch"
```

# response

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
watch=soup.select(".IRpwTa")
```

```python
watch
```

```python
watch_list=[]
```

```python
for item in watch:
    watch_list.append(item.text)
    watch_list
```
